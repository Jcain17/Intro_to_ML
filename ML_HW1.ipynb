```python
from numpy.core.multiarray import fromfile
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

file_path = '/content/drive/My Drive/D3.csv'

sample = pd.DataFrame(pd.read_csv(file_path))

X1 = sample.values[:, 0]
X2 = sample.values[:, 1]
X3 = sample.values[:, 2]
Y = sample.values[:, 3]

m = len(X2)
n = len(X1)
o = len(X3)
p = len(Y)

X_0 = np.ones((n,1))
X_0[:5]
X_1 = np.ones((m,1))
X_1[:5]
X_2 = np.ones((o,1))
X_2[:5]

X_3 = X1.reshape(n,1)
X_3[:10]
X_4 = X2.reshape(m,1)
X_4[:10]
X_5 = X3.reshape(o,1)
X_5[:10]

X1 = np.hstack((X_0, X_3))
X1[:5]
X2 = np.hstack((X_1, X_4))
X2[:5]
X3 = np.hstack((X_2, X_5))
X3[:5]

theta = np.zeros(2)

def lin_reg(X, Y, theta, m):
  predict = X.dot(theta)
  error = np.subtract(predict, Y)
  sqrError = np.square(error)
  J = 1 / (2 * m) * np.sum(sqrError)
  return J

def grad_des(X, Y, theta, alpha, inter, m):
  b = len(Y)
  hist = np.zeros(inter)

  for i in range(inter):
    predict = X.dot(theta)
    error = np.subtract(predict, Y)
    sum_delta = (alpha / b) * X.transpose().dot(error)
    theta -= sum_delta
    hist[i] = lin_reg(X, Y, theta, m)

  return theta, hist

theta1 = [0., 0.]
theta2 = [0., 0.]
theta3 = [0., 0.]
inter = 1500
alpha = 0.01

theta1, hist1 = grad_des(X1, Y, theta1, alpha, inter, m)
theta2, hist2 = grad_des(X2, Y, theta2, alpha, inter, n)
theta3, hist3 = grad_des(X3, Y, theta3, alpha, inter, o)

## Linear Regression Plots

plt.scatter(X1[:, 1], Y, color='red', marker='x', label='Traning Data')
plt.plot(X1[:, 1], X1.dot(theta1), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid(True)
plt.xlabel('X1 Values')
plt.ylabel('Y Values')
plt.title('Linear Regression for X1')
plt.legend()
plt.show()

plt.scatter(X2[:, 1], Y, color='red', marker='x', label='Traning Data')
plt.plot(X2[:, 1], X2.dot(theta2), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid(True)
plt.xlabel('X2 Values')
plt.ylabel('Y Values')
plt.title('Linear Regression for X2')
plt.legend()
plt.show()

plt.scatter(X3[:, 1], Y, color='red', marker='x', label='Traning Data')
plt.plot(X3[:, 1], X3.dot(theta3), color='green', label='Linear Regression')

plt.rcParams["figure.figsize"] = (10,6)
plt.grid(True)
plt.xlabel('X1 Values')
plt.ylabel('Y Values')
plt.title('Linear Regression for X3')
plt.legend()
plt.show()

## Gradient Decent plots

plt.plot(range(1, inter + 1), hist1, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of Iterations')
plt.ylabel('History of X1')
plt.title('Gradient Descent For X1')
plt.show()

plt.plot(range(1, inter + 1), hist2, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of Iterations')
plt.ylabel('History of X2')
plt.title('Gradient Descent For X2')
plt.show()

plt.plot(range(1, inter + 1), hist3, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of Iterations')
plt.ylabel('History of X3')
plt.title('Gradient Descentfor X3')
plt.show()

X = np.hstack((X_0, X_3, X_4, X_5))
X[:5]

theta = np.zeros(4)

theta, hist1 = grad_des(X, Y, theta, alpha, inter, m)

## Scatter Plot for X

plt.plot(range(1, inter + 1), hist1, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of Iterations')
plt.ylabel('History of X for X')
plt.title('Gradient Descent fo')
plt.show()



X1_valid = np.array([1, 1, 1, 1])
X2_valid = np.array([2, 0, 4, 3])
X3_valid = np.array([3, 2, 1, 0])

y1 = X1_valid.dot(theta)
y2 = X2_valid.dot(theta)
y3 = X3_valid.dot(theta)

print('y1 =', y1)
print('y2 =', y2)
print('y3 =', y3)
```
