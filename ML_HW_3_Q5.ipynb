```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import seaborn as sns
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.decomposition import PCA
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/cancer.csv'
dataset = pd.DataFrame(pd.read_csv(file_path))
dataset.head()
```

```python
def binmap(x):
   return x.map({'M': 1, 'B': 0})

dataset[['diagnosis']] = dataset[['diagnosis']].apply(binmap)
dataset.head()
```

```python
#Set X and Y
x_vars = np.arange(2,31,1)
X = dataset.iloc[:, x_vars].values
Y = dataset.iloc[:, 1].values

X[0:2]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

K = [5,10,15,20,25]
for k in K:
  pca = PCA(n_components=k)

  principalComponents_test = pca.fit_transform(X_test)
  principalDf_test = pd.DataFrame(data = principalComponents_test)

  principalComponents_train = pca.fit_transform(X_train)
  principalDf_train = pd.DataFrame(data = principalComponents_train)

  model = GaussianNB()

  model.fit(principalDf_train, Y_train)
  # make predictions
  expected = Y_test
  predicted = model.predict(principalDf_test)

  print("Iteration: ", k)
  print("Accuracy:",metrics.accuracy_score(expected, predicted))
  print("Precision:",metrics.precision_score(expected, predicted))
  print("Recall:",metrics.recall_score(expected, predicted))
  print("F1:",metrics.f1_score(expected, predicted))
  print('')
```
